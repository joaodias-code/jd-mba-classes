```python
df_correl = sales_df.corr()

# Generate a mask for the upper triangle
mask = np.zeros_like(df_correl, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(df_correl, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
df_correl = sales_df.corr()

for index in df_correl.index:
    df_correl.loc[(df_correl[index] < 0.7) | (df_correl[index] == 1), index] = float("NaN")

df_correl = df_correl.dropna(how='all')
df_correl = df_correl.dropna(how='all', axis=1)

df_correl.sort_values(by='OverallQual', ascending=False)


fig, ax = plt.subplots(figsize=(8, 8))
im = ax.imshow(df_correl)

# We want to show all ticks...
ax.set_xticks(np.arange(len(df_correl.index)))
ax.set_yticks(np.arange(len(df_correl.columns)))

# ... and label them with the respective list entries
ax.set_xticklabels(df_correl.index)
ax.set_yticklabels(df_correl.columns)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

im = plt.imshow(df_correl, cmap='Reds', interpolation='none')

plt.colorbar(im, shrink = 0.7)

ax.set_title("Correlação dos dados")
fig.tight_layout()
plt.show()
```
